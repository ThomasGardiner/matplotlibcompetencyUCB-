```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame


newframe = pd.merge(mouse_metadata, study_results,
                   how = 'inner', on = 'Mouse ID')


# Display the data table for preview
newframe
```

```python
# Checking the number of mice.


len(mouse_metadata)
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

data = study_results[['Mouse ID', 'Timepoint']]
occurrences = {}

for _, x in data.iterrows():
    
    if (x["Mouse ID"], x['Timepoint']) in occurrences:
        
        occurrences[(x["Mouse ID"], x['Timepoint'])] = occurrences[(x["Mouse ID"], x['Timepoint'])] + 1
        
    else:
        occurrences[(x["Mouse ID"], x['Timepoint'])] = 1

#print(occurrences)

duplicates = {x for x,y in occurrences if occurrences[(x,y)] >1}
            
print(duplicates)


```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleandf = newframe[newframe['Mouse ID'] != 'g989']
cleandf
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
#cleandf.where(cleandf['Mouse ID'] == 's185')

import matplotlib.pyplot as plt
import pandas as pd



# select an ID 
selected_mouse_id = 's185'
selected_df = cleandf[cleandf['Mouse ID'] == selected_mouse_id]


plt.plot(selected_df['Timepoint'], selected_df['Tumor Volume (mm3)'], marker='o', linestyle='-', color='b')


plt.title(f'Timepoint vs. Tumor Volume for Mouse {selected_mouse_id}')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume')
plt.grid(True)
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

capomulin_df = cleandf[cleandf['Drug Regimen'] == 'Capomulin']

# Calculate the average 'Tumor Volume (mm3)' for each unique 'Mouse ID'
average_tumor_volume = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Extract 'Weight (g)' for each unique 'Mouse ID'
weight_data = capomulin_df.groupby('Mouse ID')['Weight (g)'].first()

# Plot the scatter plot
plt.scatter(weight_data, average_tumor_volume, marker='o', color='b')

# Customize the plot
plt.title('Weight vs. Average Tumor Volume for Capomulin')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress




capomulin_df = cleandf[cleandf['Drug Regimen'] == 'Capomulin']
average_tumor_volume = capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weight_data = capomulin_df.groupby('Mouse ID')['Weight (g)'].first()



correlation_coefficient = weight_data.corr(average_tumor_volume)
slope, intercept, r_value, p_value, std_err = linregress(weight_data, average_tumor_volume)



plt.scatter(weight_data, average_tumor_volume, marker='o', color='b', label='Data points')
plt.plot(weight_data, slope * weight_data + intercept, color='r', label='Linear Regression')




plt.title('Weight vs. Average Tumor Volume for Capomulin')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.legend()
plt.grid(True)
plt.show()


```
